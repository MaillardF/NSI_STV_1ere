Une fois les données sécurisées, il est possible de les traiter normalement, cependant, un principe général est de ne jamais faire confiance à l’utilisateur final. Pour cela, en plus de sécuriser les données, Il est possible de vérifier que les variables sont bien définies et qu’elles sont éventuellement non vide !
